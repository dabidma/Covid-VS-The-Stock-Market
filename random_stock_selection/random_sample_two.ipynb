```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import random
```

```python
# Read the .csv file
company_industry = pd.read_csv('output/company_industry.csv')

# Drop all responses with duplicate tickers, N/A industries, and filter for only US Large Cap companies (Market Cap >$10B)
clean_company_industry = company_industry.drop_duplicates(subset='Ticker')
clean_company_industry = clean_company_industry.dropna(subset='Industry')
clean_company_industry = clean_company_industry.loc[(clean_company_industry['Country'] == 'US')&(clean_company_industry['Market Cap'] > 10000)]

clean_company_industry.to_csv('output/clean_company_industry.csv')
```

```python
### DO NOT RUN ###
# Randomly sample all US Large Cap companies in a given industry for only 2 companies per industry
two_stock_df = pd.DataFrame()
for industry in list(clean_company_industry.Industry.unique()):
    df_temp = clean_company_industry.loc[clean_company_industry['Industry']==industry]
    if len(df_temp) > 1:
        random_sample = random.sample(list(df_temp['Ticker']),k=2)
        df_temp = df_temp.loc[df_temp['Ticker'].isin(random_sample)]
        two_stock_df = two_stock_df.append(df_temp)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
two_stock_df
```

```python
# https://www.spglobal.com/marketintelligence/en/news-insights/blog/industries-most-and-least-impacted-by-covid-19-from-a-probability-of-default-perspective-january-2022-update
# Two random stocks from industries we are interested in observing
ten_industry_df = pd.DataFrame()
ten_industry_list = ['Airlines', 'Automobiles', 'Consumer products', 'Health Care', 'Hotels, Restaurants & Leisure', 'Logistics & Transportation', 'Pharmaceuticals', 'Real Estate', 'Telecommunication', 'Technology']
for industry in ten_industry_list:
    ten_industry_df = ten_industry_df.append(two_stock_df.loc[two_stock_df['Industry']==industry])
        
# Save the DataFrame to a .csv file
ten_industry_df.to_csv('output/ten_industry_df.csv', index=False)
```
