```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import requests
from scipy.stats import linregress
import time

# Import API Key
from api_key import api_key

# Set up client
import finnhub
fc = finnhub.Client(api_key=api_key)
```

```python
### DO NOT RUN ###
# Import .csv files of all NASDAQ and NYSE tickers and merge the two
nasdaq = pd.read_csv('input/NASDAQ.csv')
nyse = pd.read_csv('input/NYSE.csv')
merged_exchanges = nasdaq.append(nyse)

# Create empty lists to hold API responses
ticker = []
name = []
finnhubIndustry = []
country = []
marketCapitalization = []

# Gather a series of all stock tickers from the merged file
symbol_list = merged_exchanges['Symbol']

# Append all found responses to empty lists while accounting for the API call limit
for symbol in symbol_list:
    try:
        response = fc.company_profile2(symbol=symbol)
    except:
        print(f'API limit reached... Trying again in 1 minute: {symbol}')
        time.sleep(60)
        response = fc.company_profile2(symbol=symbol)
    try:
        ticker.append(response['ticker'])
        name.append(response['name'])
        finnhubIndustry.append(response['finnhubIndustry'])
        country.append(response['country'])
        marketCapitalization.append(response['marketCapitalization'])
        print(f'Loading Ticker: {symbol}')
    except KeyError:
        print(f'Error finding Ticker: {symbol}')
```

```python
# Create a DataFrame from all the API responses
company_industry = pd.DataFrame({
    'Ticker': ticker,
    'Company Name': name,
    'Industry': finnhubIndustry, 
    'Country': country,
    'Market Cap': marketCapitalization})

# Save the DataFrame to a .csv file
company_industry.to_csv('output/company_industry.csv', index=False)
```
